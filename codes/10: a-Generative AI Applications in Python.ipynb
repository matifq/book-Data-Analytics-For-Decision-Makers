```python
import gradio as gr
from transformers import pipeline

# Load a text generation pipeline
generator = pipeline('text-generation', model='gpt2')

# Define the generation function
def generate_text(prompt):
    response = generator(prompt, max_length=50, num_return_sequences=1)
    return response[0]['generated_text']

# Create Gradio Interface
gr.Interface(fn=generate_text, inputs="text", outputs="text").launch()
```

```python
from transformers import pipeline
from pyngrok import ngrok, conf
from flask import Flask, request, jsonify
import getpass

# Authentication for ngrok; for google colab
print("Enter your authtoken, which can be copied from https://dashboard.ngrok.com/get-started/your-authtoken")
conf.get_default().auth_token = getpass.getpass()

app = Flask(__name__)
port = 5000

# Initialize Text Generation Pipeline
generator = pipeline('text-generation', model='gpt2')

@app.route('/chat', methods=['GET'])
def chat():
    """
    Endpoint to interact with the chatbot.
    Example: /chat?message=Hello
    """
    print('> Ready for chatting')
    user_input = request.args.get('message')
    print("User Input:", user_input)
    if user_input is None:
        return jsonify({"response": "No input received. Use /chat?message=yourMessage"})
    else:
        response = generator(user_input, max_length=50, num_return_sequences=1)
        return jsonify({"response": response[0]['generated_text']})

@app.route('/')
def root():
    """
    Root Endpoint for Service Instructions.
    """
    return '''<h1>Welcome to the AI Chatbot!</h1>
              <p>To chat, use: <b>/chat?message=yourMessage</b></p>'''

if __name__ == "__main__":
    public_url = ngrok.connect(port).public_url
    print(f"Public URL: {public_url}")
    print("Server is running... Access via the above URL.")
    app.run(port=port)
```

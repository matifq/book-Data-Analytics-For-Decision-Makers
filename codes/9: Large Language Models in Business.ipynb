```python
# Text Summarization Example
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
# Real-World Example: NASA Mars Rover Mission (Excerpt)
text = """
NASA's Perseverance rover landed on Mars on February 18, 2021, with the primary goal of searching for signs of ancient microbial life.
It is the most advanced rover ever sent to the Red Planet and carries sophisticated scientific instruments, including the Ingenuity helicopter,
to assist in its exploration. The rover's mission focuses on astrobiology, including the collection and caching of Martian rock and soil samples.
These samples are expected to be retrieved by future missions and brought back to Earth for detailed analysis.
In addition to its scientific goals, Perseverance is testing new technologies that will benefit future human exploration of Mars,
such as oxygen production from the Martian atmosphere. The rover has already transmitted stunning images and data, offering valuable insights
into the planet's geology and climate history. Its discoveries are expected to lay the groundwork for future Mars missions and, ultimately,
human settlement on the planet.
"""
summary = summarizer(text, max_length=100, min_length=25)
print(summary[0]['summary_text'])
```

```python
from transformers import Trainer, TrainingArguments, AutoModelForSequenceClassification, AutoTokenizer
from datasets import load_dataset
import torch

# Load Dataset
dataset = load_dataset("imdb")
train_data = dataset['train'].shuffle(seed=42).select(range(5000))
val_data = dataset['test'].shuffle(seed=42).select(range(1000))

# Tokenization
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
def tokenize_function(examples):
    return tokenizer(examples['text'], truncation=True, padding='max_length')

train_dataset = train_data.map(tokenize_function, batched=True)
val_dataset = val_data.map(tokenize_function, batched=True)

# Fine-Tuning
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    evaluation_strategy="epoch"
)
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer
)
trainer.train()

# Evaluation
results = trainer.evaluate()
print(results)
```

```python
from transformers import Trainer, TrainingArguments, AutoModelForSequenceClassification, AutoTokenizer
from datasets import load_dataset
import evaluate
import numpy as np
import torch

accuracy = evaluate.load("accuracy")
# Load Dataset
dataset = load_dataset("imdb")
train_data = dataset['train'].shuffle(seed=42).select(range(500))
val_data = dataset['test'].shuffle(seed=42).select(range(100))

# Tokenization
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
def tokenize_function(examples):
    return tokenizer(examples['text'], truncation=True, padding='max_length')

train_dataset = train_data.map(tokenize_function, batched=True)
val_dataset = val_data.map(tokenize_function, batched=True)

# Fine-Tuning
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    evaluation_strategy="epoch"
)

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)

# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

# Training
trainer.train()

# Evaluation
results = trainer.evaluate()
print("Evaluation Results:")
for key, value in results.items():
    print(f"{key}: {value:.4f}")
```
